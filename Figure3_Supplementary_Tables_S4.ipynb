```r
#bed tools system call
system("bp_range=('10' '20' '30' '40')
condition=('rAF_hi' 'rAF_lo' 'sAF_hi')

df_name_1='gnomAD'
df_name_2='IGM'
df_name_3='UK.BB'

my_date=''

output_path=''

lcr_file_37='GRCh37_AllTandemRepeatsandHomopolymers_slop5.bed'
lcr_file_38='numInsteadOfChr_GRCh38_AllTandemRepeatsandHomopolymers_slop5.bed'

for n in ${condition[@]};
do 
   for i in ${bp_range[@]};
   do
      echo $i
      bedtools intersect -wa -a ${output_path}gnomad.exomes.r2.1.1.sites_indelsonly_rAF_bp${i}_${n}Indels.lt50bp.region.bed -b ${output_path}${lcr_file_37} | sort | uniq > ${output_path}${my_date}_LCR_intersect_${n}_${i}_df_1_${df_name_1}.bed
      bedtools intersect -wa -a ${output_path}2023-03-23_IGM_n39367_indelsonly_rAF_bp${i}_${n}Indels.lt50bp.region.bed -b ${output_path}${lcr_file_37} | sort | uniq > ${output_path}${my_date}_LCR_intersect_${n}_${i}_df_2_${df_name_2}.bed
      bedtools intersect -wa -a ${output_path}UK.BB.exomes.430k.sites_indelsonly_rAF_bp${i}_${n}Indels.lt50bp.region.bed -b ${output_path}${lcr_file_38} | sort | uniq > ${output_path}${my_date}_LCR_intersect_${n}_${i}_df_3_${df_name_3}.bed
   done;
done", intern = TRUE)
```

# Calculating Cramer's V 

```r
df = c(1,2,3)
bp_range = c("10", "20", "30", "40")

for (i in df){
  for (k in bp_range) {
    temp_rAF_hi_LCR = paste0("rAF_hi_LCR_bp_", k, "_df_", i )
    temp_rAF_hi_not_LCR = paste0("rAF_hi_not_in_LCR_bp_", k, "_df_",i)
      
    temp_rAF_lo_LCR = paste0("rAF_lo_LCR_bp_", k, "_df_", i)
    temp_rAF_lo_not_LCR = paste0("rAF_lo_not_in_LCR_bp_", k, "_df_", i)
      
    temp_sAF_hi_LCR = paste0("sAF_hi_LCR_bp_", k, "_df_", i)
    temp_sAF_hi_not_LCR = paste0("sAF_hi_not_in_LCR_bp_", k, "_df_", i)
    
    temp_cramerV_rAFhi_sAFhi = paste0("cramerV_rAFhi_sAFhi_bp_", k, "_df_", i)
    temp_cramerV_rAFhi_rAFlo = paste0("cramerV_rAFhi_rAFlo_bp_", k, "_df_", i)
    
    data_rAFhi_sAFhi = matrix(c(get(temp_rAF_hi_LCR), get(temp_rAF_hi_not_LCR), get(temp_sAF_hi_LCR), get(temp_sAF_hi_not_LCR)), nrow = 2)
    
    data_rAFhi_rAFlo = matrix(c(get(temp_rAF_hi_LCR), get(temp_rAF_hi_not_LCR), get(temp_rAF_lo_LCR), get(temp_rAF_lo_not_LCR)), nrow = 2)
    
    cramerV_rAFhi_sAFhi = cramerV(data_rAFhi_sAFhi)
    cramerV_rAFhi_rAFlo = cramerV(data_rAFhi_rAFlo)
    
    assign(temp_cramerV_rAFhi_sAFhi, cramerV_rAFhi_sAFhi)
    assign(temp_cramerV_rAFhi_rAFlo, cramerV_rAFhi_rAFlo)
    
  }
}

## creating a table for Cramer's V

## gnomAD

df1_rAFhi_sAFhi = c(cramerV_rAFhi_sAFhi_bp_10_df_1, cramerV_rAFhi_sAFhi_bp_20_df_1, cramerV_rAFhi_sAFhi_bp_30_df_1, cramerV_rAFhi_sAFhi_bp_40_df_1)
df1_rAFhi_rAFlo = c(cramerV_rAFhi_rAFlo_bp_10_df_1, cramerV_rAFhi_rAFlo_bp_20_df_1, cramerV_rAFhi_rAFlo_bp_30_df_1,cramerV_rAFhi_rAFlo_bp_40_df_1)

df1_cramerV = rbind (bp_range, df1_rAFhi_sAFhi, df1_rAFhi_rAFlo)
df1_cramerV = as.data.frame(df1_cramerV)
df1_cramerV
fwrite(df1_cramerV, "df1_cramerV.csv", row.names = T)


## IGM

df2_rAFhi_sAFhi = c(cramerV_rAFhi_sAFhi_bp_10_df_2, cramerV_rAFhi_sAFhi_bp_20_df_2, cramerV_rAFhi_sAFhi_bp_30_df_2, cramerV_rAFhi_sAFhi_bp_40_df_2)
df2_rAFhi_rAFlo = c(cramerV_rAFhi_rAFlo_bp_10_df_2, cramerV_rAFhi_rAFlo_bp_20_df_2, cramerV_rAFhi_rAFlo_bp_30_df_2,cramerV_rAFhi_rAFlo_bp_40_df_2)

df2_cramerV = rbind (bp_range, df2_rAFhi_sAFhi, df2_rAFhi_rAFlo)
df2_cramerV = as.data.frame(df2_cramerV)
df2_cramerV
fwrite(df2_cramerV, "df2_cramerV.csv", row.names = T)

## UKBB

df3_rAFhi_sAFhi = c(cramerV_rAFhi_sAFhi_bp_10_df_3, cramerV_rAFhi_sAFhi_bp_20_df_3, cramerV_rAFhi_sAFhi_bp_30_df_3, cramerV_rAFhi_sAFhi_bp_40_df_3)
df3_rAFhi_rAFlo = c(cramerV_rAFhi_rAFlo_bp_10_df_3, cramerV_rAFhi_rAFlo_bp_20_df_3, cramerV_rAFhi_rAFlo_bp_30_df_3,cramerV_rAFhi_rAFlo_bp_40_df_3)

df3_cramerV = rbind (bp_range, df3_rAFhi_sAFhi, df3_rAFhi_rAFlo)
df3_cramerV = as.data.frame(df3_cramerV )
df3_cramerV 
fwrite(df3_cramerV, "df3_cramerV.csv", row.names = T)
```
